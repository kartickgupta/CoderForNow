```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import mglearn
```

```python
iris_dataset = load_iris()
iris_dataset.keys()
```

```python
print(iris_dataset['DESCR'])
```

```python
iris_df = pd.DataFrame(iris_dataset['data'])
```

```python
iris_df.rename(columns = {0: 'P. Length', 1: 'P Width', 2: 'S. Length', 3: 's. Width'}, inplace=True)
iris_df.describe()
```

```python
X = np.array(iris_df)
y = iris_dataset['target']
```

#### Before making the split, the train_test_split function shuffles the dataset using a pseudorandom number generator

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.25)
```

```python
print(X_train.shape, y_train.shape)
```

```python
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset['feature_names'])
```

```python
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(25, 25), marker='o',
hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
knn = KNeighborsClassifier(n_neighbors = 1,n_jobs = -1)
```

```python
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
prediction_data = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(prediction_data)
print(prediction)
```

```python
print(f'Prediction : {iris_dataset.target_names[prediction]} ')
```
