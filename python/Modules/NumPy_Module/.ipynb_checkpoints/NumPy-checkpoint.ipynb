```python
import numpy as np
import matplotlib.pyplot as plt
```

# Declaring a Numpy Array

#### Create an array, each element of which is zero with data type = float64

```python
z = np.zeros((9,2)) # Makes an array of 9 x 2
print(type(z[0][0]))
z
```

#### Create an array, each element of which is 1 with data type = float64

```python
z = np.ones((6,5), np.int32)
print(type(z[0][0]))
z
```

#### Number of dimensions in the array

```python
print(z.ndim)
z.shape
```

#### shape : Returns =  tuple of ints,  (height, width, depth)

```python
z.shape = (10, 3)
z
```

#### Create an array, but leave its allocated memory unchanged (i.e., it contains "garbage").

```python
z = np.empty((4,5)) 
z
```

#### Returns an array of numbers (c) from a to b inclusive

```python
z = np.linspace(2, 10, 50) # get a single array 50 nos. from 2 to 10 inclusive
z
```

#### Construct an array.

```python
z = np.array([1, 3, 5, 6, 7,8])
z
```

```python
a_list = [34, 45, 64,65, 6,21, 132, 904]
z = np.array([a_list, a_list])
print(z.shape)
print(z.ndim)
z
```

#### .size - Number of elements in the array.

```python
print(z)
z.size
```

# Statistical Operations

```python
image_sin = np.sin(img) # Applies sine function to all the values in the array
image_sin
```

```python
print(np.sum(img))
print(np.product(img))
print(np.max(img))
print(np.min(img))
print(np.mean(img))
print(np.std(img))
print(np.argmax(img))
print(np.argmin(img))
```
