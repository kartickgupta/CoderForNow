```python
df = pd.read_csv('avocado.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace = True)
df.sort_index(ascending = True, inplace = True)
df = df[df['type'] == 'organic']
```

```python
group_df = pd.DataFrame()
for region in df['region'].unique():
    print(region)
    region_df = pd.DataFrame()
    region_df = df.copy()[df['region'] == region]
    region_df[f'{region}_25ma'] = region_df['AveragePrice'].rolling(25).mean() # rolling makes a window
    
    if group_df.empty:
        group_df = region_df[[f'{region}_25ma']] # Copy DataFrame
    else:
        group_df = group_df.join(region_df[[f'{region}_25ma']]) # We add columns(series) in the DataFrame
```
